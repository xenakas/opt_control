```r
df_ihh  <-  read.csv(file="raw_data.csv", header=TRUE, sep=",", row.names =1)
```

```r
df_ihh$MA[df_ihh$MA == 0] <- NA
```

```r
library('plm')
```

```r
df_ihh = pdata.frame(df_ihh, index = c('otrasl','year'))
```

```r
df_ihh
```

```r
df_ihh['ma_lag'] = lag(df_ihh$MA, 1)
```

```r
df_lag = lag(df_ihh$MA, 1)
```

```r
df_ihh['log_ihh'] = log(df_ihh$IHH)
df_ihh['log_ma'] = log(df_ihh$MA)
```

```r
coplot(log_ihh ~ year|otrasl, type="b", data=df_ihh)
```

```r
library(gplots)
```

```r
plotmeans(log_ihh ~ year, data = df_ihh)
# plotmeans(IHH ~ otrasl, data = df_ihh)
```

```r
ols <-lm(log_ihh~log_ma, data = df_ihh)
summary(ols)
```

```r
library(ggplot2)
```

```r
yhat <- ols$fitted
ggplot(df_ihh, aes(x = MA, y = log_ihh))+
  geom_point() 
# +
#    geom_smooth(method=lm)
```

```r
dim(df_ihh)
```

```r
# na.omit(df_ihh)
df_ihh_omit = df_ihh[rowSums(is.na(df_ihh)) <= 1,]
```

```r
fixed.dum <-lm(log_ihh~MA + factor(otrasl) - 1, data = df_ihh)
summary(fixed.dum)
```

```r
library(car)  
```

```r
yhat <- fixed.dum$fitted
scatterplot(yhat ~ df_ihh_omit$MA | df_ihh_omit$otrasl,  xlab ="x1", ylab ="yhat", boxplots = FALSE,smooth = FALSE)
abline(lm(df_ihh_omit$log_ihh~df_ihh_omit$MA), lwd=3, col="red")
```

```r
fixed <- plm(log_ihh~log_ma, data = df_ihh, model="within")
summary(fixed)
```

```r
fixef(fixed)
```

```r
pFtest(fixed, ols)
```

p-value < 0.05 выбираем fe model 

```r
random <- plm(log_ihh~log_ma, data = df_ihh, model="random")
summary(random)
```

```r
phtest(fixed, random)
```

p-value > 0.5 используем  RE модель

```r
fixed.time <- plm(log_ihh~log_ma + factor(year), data=df_ihh, model="within")
summary(fixed.time)

```

```r
pFtest(fixed.time, fixed)
```

```r
plmtest(fixed, c("time"), type=("bp"))
```

Есть гетероскдастичность, необходимо использовать устойчивые к гетероскедастичности оценки ковариационной матрицы 

```r
coeftest(random) 
```

```r
coeftest(random, vcovHC) 
```

```r
coeftest(random, vcovHC(random, type = "HC3")) 
```
