```r
df_ihh  <-  read.csv(file="raw_data.csv", header=TRUE, sep=",", row.names =1)
```

```r
library('plm')
```

```r
df_ihh = pdata.frame(df_ihh, index = c('otrasl','year'))
```

```r
ols <-lm(IHH~MA, data = df_ihh)
summary(ols)
```

```r
library(ggplot2)
```

```r
yhat <- ols$fitted
ggplot(df_ihh, aes(x = MA, y = IHH))+
  geom_point() 
# +
#    geom_smooth(method=lm)
```

```r
fixed.dum <-lm(IHH~MA + factor(otrasl) - 1, data = df_ihh)
summary(fixed.dum)
```

```r
library(car)  
```

```r
yhat <- fixed.dum$fitted
scatterplot(yhat ~ df_ihh$MA | df_ihh$otrasl,  xlab ="x1", ylab ="yhat", boxplots = FALSE,smooth = FALSE)
abline(lm(df_ihh$IHH~df_ihh$MA),lwd=3, col="red")
```

```r
fixed <- plm(IHH~MA, data = df_ihh, model="within")
summary(fixed)
```

```r
fixef(fixed)
```

```r
pFtest(fixed, ols)
```

```r
random <- plm(IHH~MA, data = df_ihh, model="random")
summary(random)
```

```r
phtest(fixed, random)
```

    - (IHH<1000) &  (IHH>=400)
    - (IHH<400) &  (IHH>=200)
    - (IHH<200) &  (IHH>=100)
