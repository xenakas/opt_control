```r
library(sophisthse)
library(ggplot2)
library(forecast)
library(uroot)
library(urca)
```

```r
library(lmtest)
library(dplyr)
library(broom)
library(strucchange)
library(tseries)
library(readxl)
library(TSA)
library(car)
library(zoo)
library(vars)
```

```r
gdp1 <- sophisthse("GDP_Q_I") # скачаем Индекс реального ВВП (GDP_Q_DIRI) с 2000.1 по 2002.4
gdp2 <- sophisthse("GDPEA_Q") # скачаем Индекс реального ВВП согласно ОКВЭД (GDPEA_Q_DIRI) с 2003.1 по 2019.1 
gdp <-  ts(data = c(gdp1[25:36, "GDP_Q_DIRI"], gdp2[1:65, "GDPEA_Q_DIRI"]), start = c(2000,1), frequency = 4 )  # собрали в один ряд 
```

```r
ggtsdisplay(gdp, main="") 
```

```r
inv1 <- sophisthse("INVFC_Q") # скачаем Инвестиции в основной капитал (INVFC_Q)  с 2000.1 по 2019.1 
inv <- ts(data = inv1[29:105] , start = c(2000,1), frequency = 4) 
```

```r
ggtsdisplay(inv, main="") 
```

```r
lab1 <- sophisthse("EMPLDEC_M")  # скачаем Заявленная потребность в работниках (на конец месяца) (EMPLDEC_M) с 2000.1 по 2009.3
lab <- ts(data =   lab1[73:303] , start = c(2000,1), frequency = 12 ) 
```

```r
ggtsdisplay(lab, main="")
```

```r
# проведем тест Дики-Фуллера на единичный корень
summary(ur.df(gdp)) 
summary(ur.df(inv))
summary(ur.df(lab))
```

```r
# проведем тест Дики-Фуллера на единичный корень
summary(ur.df(gdp, type='trend', lags = 4,  selectlags = "BIC")) 
summary(ur.df(inv, type='trend', lags = 4, selectlags = "BIC"))
summary(ur.df(lab, type='trend', lags = 12, selectlags = "BIC"))
```

```r
gdp_diff <- diff(gdp, 4)
ggtsdisplay(gdp_diff)
ggseasonplot(gdp_diff)  
```

```r
model_gdp <- auto.arima(gdp) 
summary(model_gdp)
```

```r
checkresiduals(model_gdp)
```

```r
summary(ur.df(gdp_diff, type='trend', lags = 4,  selectlags = "BIC")) 
```

```r
model_gdp <- Arima(gdp, order = c(1,0,2), seasonal = c(0,1,0),  include.drift=T,  method = "ML")
summary(model_gdp)
```

```r
checkresiduals(model_gdp)
```

```r
future_gdp <- forecast(model_gdp, h = 8)
future_gdp
```

```r
autoplot(future_gdp)
```
