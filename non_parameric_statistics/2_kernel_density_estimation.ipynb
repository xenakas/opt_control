## Kernel Density Estimation

```r
head(faithful)
```

```r
x = faithful$eruptions
```

```r
options(repr.plot.width=5, repr.plot.height=4)
```

```r
hist(x, breaks = 50)
```

```r
plot(density(x))
polygon(density(x), col = 'wheat')
```

```r
K  <-  eval(formals(density.default)$kernel)
```

### Kernel Comparison

```r
plot(density(x))
for (i in 2:7){
    lines(density(x,kernel=K[i]), col=i)
}
legend(4.5,0.25, legend=K, col=1:7, cex=0.6, lty=1, y.intersp = 1.7)
```

All kernels give approximately the same level of efficiency:

```r
dens_list  <-  c()
for (j in 1:7){
    dens_list  <- c(dens_list, density(kernel=K[j], give.Rkern=T))
    }
dens_list
K
efficiencies  <-  dens_list[2]/dens_list
efficiencies
```

### Bandwidth Comparison (for gaussian kernel)

```r
band = c("nrd","nrd0", "ucv", "bcv", "SJ")
plot(density(x,bw=band[1]), ylim = c(0, 0.65))
for (i in 2:5){
    lines(density(x, bw=band[i]), col=i)
}
legend(5.5,0.25, legend=band, col=1:5, cex=0.7, lty=1, y.intersp = 1.7)
```
