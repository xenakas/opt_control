### Simulating from Bart Simpson Distribution

$$p(x) = 1/2 p_{(0,1)} + 1/10 \sum^4_{j=0} p_{(j/2-1,1/10)}(x)$$ 


$p_(\mu, \sigma)$ - density of $N(\mu, \sigma)$

$p_1,...,p_k$ 

$a_1+...+a_k=1, a_i\geq0$

$\sum_{i=j}^k a_i p_i(x)$ <- probability density (mixtures) 

can't just sum (because to get pdf of sum of rv we need convolution)

#### Simulating from given distribution:

$\eta = 1,...,6$ with p $1/2, 1/10, ... ,1/10$

```r
n = 1000
eta = sample(1:6, size=n, prob=c(0.5,rep(0.1,5)), replace = TRUE)
table(eta)
```

$\mu_1 =0, \mu_2 = -1, \mu_3 = -1/2, ..., \mu_6 =1 $

```r
m = c(0,-1,-0.5,0,0.5,1)
```

$\sigma_1^2 = 1, \sigma_2^2 = 1/10 = ... =  \sigma_6^2 = 1/10$

```r
sigma = c(1,rep(0.1,5))
```

$X \sim N(\mu_k, \sigma_k^2)$ if  $\eta=k$

```r
x=rep(0,n)

for(j in 1:n){
    x[j] = rnorm(1, mean=m[eta[j]], sd = sigma[eta[j]])
}
```

```r
options(repr.plot.width=5, repr.plot.height=4)
```

```r
hist(x)
```

```r
hist(x, breaks = 100)
```

#### Estimating $p(x)$ using histogram

##### Choice of h by Scott

The default for breaks is "Sturges": see nclass.Sturges. Other names for which algorithms are supplied are "Scott" and "FD" / "Freedman-Diaconis" (with corresponding functions nclass.scott and nclass.FD).

by Scott's rule approximate amount of bins (doesn't work because of c command pretty):

$$h_{opt} = \frac{1}{n^{1/3}} * \left( \frac{6}{\int(p'(u))^2 du} \right)$$

Scott: $p \sim N(0,\sigma^2)$

$h_{opt} = \frac{1}{n^{1/3}} * \left( 24 * \sqrt{...} \right)^{1/3} \sigma = \frac{1}{n^{1/3}}  * 3.5 * \sigma $


```r
scott_const = ceiling(6/(3.5 * sd(x) / n^(1/3)))
scott_const
```

```r
pretty(x,20)
```

```r
hist(x, breaks = 'Scott')
```

```r
hist(x, breaks = seq(min(x),max(x),l = ceiling(scott_const)))
```

##### Choice of h by Freedman-Diaconis (1926)

$$h_{opt} = \frac{1}{n^{1/3}} * \left( \frac{6}{\int(p'(u))^2 du} \right)$$

FD: $h = 2 * IQR * 1/n^{1/3}$


```r
H  <-  hist(x, breaks = 'FD')
```

##### Choice of h by Sturges (1926)

$m = \lceil log_2 n \rceil$

```r
sturges_const = ceiling(log(n, base=2))
sturges_const
```

```r
hist(x, breaks = 'Sturges')
```

```r
hist(x, breaks = seq(min(x),max(x),l = ceiling(sturges_const)))
```

#### We know actual distribution


```r
p = function(x){
    y = dnorm(x)/2
    for(j in (0:4)){
        y = y + 0.1*dnorm(x,mean=m[j+2],sd=sigma[j+2])
    }
    return(y)
}
```

```r
u = seq(length = n, from = -2, to =2)
plot(u,sapply(u,p), type = "l", col = "red")
```

```r
plot(H, freq = F, ylim = c(0, max(sapply(u,p))))
points(u,sapply(u,p), type = "l", col = "red")
```

```r
pn = function(x){
    H$density[max(which(H$breaks<x))]
    }
```

```r
mse = function(x){
    return((pn(x)-p(x))^2)
}
```

```r
res = rep(0, 100)
for (br in (1:100)){
    H = hist(x, breaks = br, plot = F)
    res[br] = sum(sapply(u, mse))
}
```

```r
which.min(res)
```

```r
H = hist(x, breaks = 40, plot=F)
plot(H, freq = F,  ylim = c(0, max(sapply(u,p)+0.1)))
points(u, sapply(u,p), type = "l", col= "red")
```
