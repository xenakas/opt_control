```r
install.packages("survival")
install.packages("survminer")
library(survival)
library(survminer)
```

```r
data("lung")
```

```r
head(lung)
summary(lung)
```

```r
agecat=rep(1,length(lung$age))
agecat[which(lung$age>50 & lung$age<=70)]=2
agecat[lung$age>70]=3
```

```r
t=table(agecat, lung$ph.ecog)
t
chisq.test(t)
chisq.test(t[,1:3])
chisq.test(t[,1:3], simulate.p.value = TRUE)
```

```r
t=table(lung$status, lung$sex)
t
fisher.test(t)
```

```r
s=survfit(Surv(time,status)~1, data=lung)
plot(s)
```

```r
library(ggplot2)
s=survfit(Surv(time,status)~sex, data=lung)
ggsurvplot(s,
           conf.int=TRUE,
           ggtheme=theme_light(),
           legend.labs=c("male","female"),
           xlab="days",
             risk.table = TRUE)
```

```r
h=lung$time[lung$status==2]
d=density(h)          
plot(d)
```

```r
res=rep(0, length(s$time))
for (i in (1: length(s$time))){
  res[i]=
    d$y[which.min(abs(d$x-s$time[i]))]
}
r=res/s$surv
plot(s$time,r, type="l")
```
