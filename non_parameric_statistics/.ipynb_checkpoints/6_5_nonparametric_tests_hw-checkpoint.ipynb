### N1

Let a r.v. $X \sim U[0,1]$. Consider three models:

1. monotone dependence: $Y = \sin(\pi X/2)$;
2. monotone dependence with small noise: $Y = \sin(\pi X/2) + e$, where
$ e \sim N(0,0.04)$;
3. non-monotone dependence with small noise: $Y = \sin(\pi X) + e$.

The general aim of the tasks listed below is to study whether the analysis of various correlation coefficients can ”catch” various types of dependence. Also these tasks illustrate the difference between various methods.

#### Simulate samples $(X_1, Y_1), ...(X_n, Y_n)$ from each model $M$ times.

For each model calculate
the Pearson, Kendall and Spearman correlation coefficients and the
corresponding p-values. If for some model and some method you
get essentially different p-values, display the boxplot.

```r
n  <- 1000
m  <- 20
corrs  <-  c("pearson", "kendall", "spearman")
pvalues  <- data.frame(matrix(NA, nrow = m, ncol = 9))
cors  <- data.frame(matrix(NA, nrow = m, ncol = 9))
colnames(cors) <- colnames(pvalues)  <- c(rep(corrs,3))
```

```r
for (k in 1:3) {
    for (j in 1:3) {
        pval <- c()
        cor <- c()
        for (i in 1:m) {
            x <- runif(n, 0, 1)
            e <- rnorm(n, 0, 0.2)
            if (k == 1) {
                y <- sin(pi * x/2)
            } else if (k == 2) {
                y <- sin(pi * x/2) + e
            } else {
                y <- sin(pi * x) + e
            }
            test <- cor.test(x, y, method = corrs[j])
            pval <- c(pval, test$p.value)
            cor <- c(cor, test$estimate)
        }
        pvalues[, 3 * (k - 1) + j] <- pval
        cors[, 3 * (k - 1) + j] <- cor
    }
}
```

```r
boxplot(cors)
boxplot(pvalues)
```

#### In the model (b), increase the variance of the noise.
For each $\sigma^2$, calculate the Kendall and Spearman correlation coefficients. Find the maximal $\sigma^2$, such that the
corresponding correlation coefficients are still significant.

```r
sigma_sq = seq(0.5, 15, by = 0.5)
k  <-  2
cn <- 1

for (j in 2:3) {
    for (l in 1:length(sigma_sq)) {
        pval <- c()
        cor <- c()
        for (i in 1:m) {
            x <- runif(n, 0, 1)
            e <- rnorm(n, 0, sqrt(sigma_sq[l]))
            if (k == 1) {
                y <- sin(pi * x/2)
            } else if (k == 2) {
                y <- sin(pi * x/2) + e
            } else {
                y <- sin(pi * x) + e
            }
            test <- cor.test(x, y, method = corrs[j])
            pval <- c(pval, test$p.value)
            cor <- c(cor, test$estimate)
        }
        pvalues[, cn] <- pval
        cors[, cn] <- cor
        cn <- cn + 1
    }
}
colnames(cors) <- colnames(pvalues)  <- c(rep(sigma_sq,2))
```

```r
boxplot(cors)
boxplot(pvalues)
```

Spearman (1-% significance level)

```r
which(colMeans(pvalues)[(length(sigma_sq)+1):(2*length(sigma_sq))]>=0.01)
```

Kendall (5-% significance level)

```r
which(colMeans(pvalues)[1:length(sigma_sq)]>=0.05)
```

Spearman (5-% significance level)

```r
which(colMeans(pvalues)[(length(sigma_sq)+1):(2*length(sigma_sq))]>=0.05)
```

```r
which(colMeans(pvalues)[1:length(sigma_sq)]>=0.1)
```

Spearman (10-% significance level)

```r
which(colMeans(pvalues)[(length(sigma_sq)+1):(2*length(sigma_sq))]>=0.1)
```

#### For $\sigma^2$ obtained on the previous step (5 %), compare Kendall and Spearman correlation coefficients. 

Test the hypothesis that the location shift (median of their difference) is equal to zero. Calculate the corresponding confidence interval.

```r
corrs  <-  c("pearson", "kendall", "spearman")
cors  <- data.frame(matrix(NA, nrow = m, ncol = 2))
colnames(cors)  <-  corrs[2:3]

sgms <- c(sigma_sq[which(colMeans(pvalues)[1:length(sigma_sq)] >= 0.05)[1]], sigma_sq[which(colMeans(pvalues)[(length(sigma_sq) + 
    1):(2 * length(sigma_sq))] >= 0.05)[1]])

for (j in 2:3) {
        pval <- c()
        cor <- c()
        for (i in 1:m) {
            x <- runif(n, 0, 1)
            e <- rnorm(n, 0, sqrt(sgms[j-1]))
            if (k == 1) {
                y <- sin(pi * x/2)
            } else if (k == 2) {
                y <- sin(pi * x/2) + e
            } else {
                y <- sin(pi * x) + e
            }
            test <- cor.test(x, y, method = corrs[j])
            pval <- c(pval, test$p.value)
            cor <- c(cor, test$estimate)
        }
        cors[, j-1] <- cor
    }
```

```r
boxplot(cors)
```

```r
wilcox.test(cors[,1], cors[,2], paired = TRUE, conf.int = T, conf.level = 0.05)
```

#### Analyse more precisely fertility rates in the first 10 provinces. 

For
each of these provinces, find a province in the rest database (that is,
among observations from 11 to 47) with the closest social-economical
indicators. As the measure of closeness use the Euclidian distance
between scaled variables. Test the hypothesis that fertility rates in
the first 10 provinces are the same as in the provinces with similar
socio-economical indicators. Interpret the results.

```r
data <- scale(swiss[, 2:6], center = FALSE, scale = colSums(swiss[, 2:6]))
res <- c()
for (i in (1:10)) {
    x <- data[i, ]
    min <- 10000
    ind <- 1000
    for (j in 11:47) {
        m <- dist(rbind(x, data[j, ]))
        if (m < min) {
            min <- m
            ind <- j
        }
    }
    res[i] <- ind
}
```

```r
wilcox.test(data[1:10,1],data[res,1], paired=T, alternative = "two.sided")
```

Fertility rates in the provinces with similar socio-economical indicators are the same. 


####   Divide all provinces into 3 groups, which we will denote by C,P,M:

- C - more than 80 % of population are Catholics; 
- P - more than 80 % are Protestants; 
- M- ”mixed” provinces (more or equal than 20 % Catholics, and more or equal than 20 % Protestants). 

```r
C <- swiss[which(swiss[, 5] > 80), 1]
P <- swiss[which(swiss[, 5] < 20), 1]
M <- swiss[which((swiss[, 5] >= 20) & (swiss[, 5] <= 80)), 1]
```

```r
boxplot(C, P, M, names=c('Catholic', 'Protestant', 'Mixed'))
```

Test
the hypothesis that in all of these groups the fertility rate follows
the same distribution.

```r
kruskal.test(list(C, P, M))
```

Null
     that the location parameters of the distribution of ‘x’ are the
     same in each group (sample) is rejected.   The alternative is that they differ
     in at least one.

Consider also similar hypothesis pairwise
(that is, for C and P group, C and M, P and M), and use the most
appropriate alternative for each pair. 

```r
wilcox.test(C, P, alternative = 'greater') 
```

```r
wilcox.test(C, M, alternative = 'greater')
```

```r
wilcox.test(P, M)
wilcox.test(P, M, alternative = 'greater')
wilcox.test(P, M, alternative = 'less')
```

For C fertility rates are greater that for P. For C fertility rates are greater that for M. There i not enough evidence that fertility rates for P and M differ.

#### Within each group obtained on the previous step, divide all provinces into 4 subgroups:
 
- with more than 50% of males involved in agriculture, and small value of infant mortality (say, less than 1-st quartile of the infant mortality across the country);
- with less than 50% of males involved in agriculture, and small value of infant mortality;
- with more than 50% of males involved in agriculture, and high value of infant mortality (more than 1-st quartile of the infant mortality across the country);
- with less than 50% of males involved in agriculture, and high value of infant mortality.


Compute the average value of Fertility for each subgroup. Test the
hypothesis that the medians of the distributions of average Fertilities (averaging among subgroups 1-4) are the same for groups C,
P, M. Interpret the results.

```r
head(swiss)
```

```r
avg  <-  c()
for (i in 1:3){
    if (i==1){
        dat = swiss[which(swiss[, 5] > 80), ]
    }else if (i==2){
        dat =  swiss[which(swiss[, 5] < 20), ]
    }else{
        dat = swiss[which((swiss[, 5] >= 20) & (swiss[, 5] <= 60)), ]
    }
    
    q1 = quantile(dat[, 6], 0.25)
    q2 = quantile(dat[, 2], 0.5)
        
    avg  <-  c(avg, 
               mean(dat[(dat[,6]<q1)& (dat[,2]>q2), 1]),
               mean(dat[(dat[,6]<q1)& (dat[,2]<q2), 1]),
               mean(dat[(dat[,6]>q1)& (dat[,2]>q2), 1]),
               mean(dat[(dat[,6]>q1)& (dat[,2]<q2), 1])
              )
}
```

```r
friedman.test(matrix(avg, nrow=4))
```

The null hypothesis  that apart from an effect of ‘blocks’, the    location parameter of ‘y’ is the same in each of the ‘groups’ cannot be rejected, medians in each groups are the same.
