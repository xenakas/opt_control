```r
x  <- cars$speed
y  <-  cars$dist
plot(x,y)
```

```r
x_un  <-  unique(x)
y_un <-  sapply(unique(x), function(z){ mean(y[which(x==z)])})
plot(x_un, y_un)
```

```r
install.packages('orthopolynom')
```

```r
library('orthopolynom')
```

```r
summary(x)
summary(x_un)
```

```r
A = min(x_un)
B = max(x_un)
x2  = 2*(x_un-A)/(B-A) - 1
```

```r
plot(x2)
```

```r
plot(x_un,y_un, ylim=c(-15, 130))
er  <- c()
for (J in 1:5){
    LP = legendre.polynomials(J, normalized = T)
    leg = matrix(unlist(polynomial.values(polynomials = LP, x = x2)), ncol=J+1, byrow=F)
    coef = colMeans(leg*y_un)
    res = 2*leg%*%coef
    lines(x_un, res, col=J)
    er <- c(er, mean((res-y_un)^2))
}
```

```r
f = function(x){
    sqrt(x*(1-x))*sin(2.1*pi/(x+0.05))
}
x  <-  seq(0,1,length = 2^10)
y  <- sapply(x,f)
plot(x,sapply(x,f), type = 'l')
```

```r
install.packages("waveslim")
```

```r
library('waveslim')
```

```r
m = mra(y, method = "dwt", wf = 'haar', J=10)
plot(m[[8]], type = 'l')
lines(m[[9]], col = 4)
lines(m[[10]], col = 2)
lines(m[[11]], col = 3)
```

```r
res= rep(0, length(m[[11]]))
for (j in 11:1){
    res=res+m[[j]]
    plot(res,main=j, type='l')
#     readline(prompt = "Press enter")
}

```
